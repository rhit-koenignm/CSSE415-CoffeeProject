## After exploring github a bit, I discovered two datasets containing arabica and robusta data with the same number of features and type of features. ### Now to rename some comments I notice are different, but first lets check that the values are similar enough

```python
# First difference is 'Aroma' and 'Fragrance...Aroma' so lets check the values and compare
arabica_cleaned['Aroma'].value_counts()
```

```python
robusta_cleaned['Fragrance...Aroma'].value_counts()
```

```python
# Since it is clear that these two types are the same, I'll change robusta's column to be 'Aroma'
robusta_cleaned = robusta_cleaned.rename(columns={"Fragrance...Aroma": "Aroma"})
robusta_cleaned.columns
```

```python
arabica_cleaned.columns
```

### Aroma has now been fixed, time for acidity!

```python
arabica_cleaned['Acidity'].value_counts()
```

```python
robusta_cleaned['Salt...Acid'].value_counts()
```

```python
# Once again, I'd say these are on the same scale and value so I'll rename the robusta's to match
robusta_cleaned = robusta_cleaned.rename(columns={'Salt...Acid': 'Acidity'})
robusta_cleaned.columns
```

### Acidity fixed! Now for sweetness

```python
arabica_cleaned['Sweetness'].value_counts()
```

```python
robusta_cleaned['Bitter...Sweet'].value_counts()
```

```python
# Since it appears that the second one is actually a rating on 0 to 10 where 0 is bitter and 10 is sweet, 
# I think it is safe to rename the column in robusta so we can combine
robusta_cleaned = robusta_cleaned.rename(columns={'Bitter...Sweet': 'Sweetness'})
robusta_cleaned.columns
```

### Done with Sweetness, now onto Mouthfeel!

### Since there was no 'Mouthfeel' feature within the arabica dataset, so I did a bit of research. When dealing with coffee, the term 'body' refers to the mouthfeel of the coffee, so I'll compare  arabica's 'Body' column with robusta's 'Mouthfeel'

```python
arabica_cleaned.Body.value_counts()
```

```python
robusta_cleaned.Mouthfeel.value_counts()
```

### Looking at these values, it looks like body and mouthfeel are synonymous and scaled the same, so it should be safe to combine the two.

```python
robusta_cleaned = robusta_cleaned.rename(columns={'Mouthfeel': 'Body'})
robusta_cleaned.columns
```

```python
arabica_cleaned.columns
```

### Next lets take a look at Uniformity vs Uniform.Cup

```python
robusta_cleaned['Uniform.Cup'].value_counts()
```

### So I think it is safe to assume that Uniform.Cup and Uniformity are the same, so we'll change robusta to match.

### The next bunch of columns (Clean.Cup, Balance, Cupper.Points,Total.Cup.Points, Moisture, Category.One.Defects, Quakers, Color, Category.Two.Defects, Expiration, Certification.Body, Certification.Address, Certification.Contact, unit_of_measurement, altitude_low_meters, altitude_high_meters, and altitude_mean_meters) are named the same so we should be able to merge the two together once we rearrange the columns to be in the same order.

## Rearranging columns to prep for merging

```python
# Let's look at the current ordering of the arabica dataset
arabica_cleaned.columns
```

```python
robusta_cleaned.columns
```

```python
robusta_cols = robusta_cleaned.columns.tolist()
robusta_cols
```

```python
arabica_cleaned.columns.tolist()
```

```python
robusta_cols = robusta_cols[0:24] + robusta_cols[25:26] + robusta_cols[28:29] + robusta_cols[26:28] + robusta_cols[24:25] + robusta_cols[29:]
robusta_cols
```

```python
# They are now in the same order, so we are good to merge the two!
ordered_robusta = robusta_cleaned.reindex(robusta_cols, axis=1)
ordered_robusta.columns.tolist()
```

```python
merged_coffee = pd.concat([arabica_cleaned, ordered_robusta])
merged_coffee
```

```python
merged_coffee.to_csv('./merged_coffee.csv', index=False)
```

## Now let's chop off some of those features!!

```python
# Goodbye unnamed column, row numbers are good enough
merged_cleaned = merged_coffee.drop('Unnamed: 0', axis=1)
merged_cleaned.head(3)
```

```python
# Since owner can be extremely specific, probably best to remove this one for now
merged_cleaned = merged_cleaned.drop('Owner', axis=1)
merged_cleaned.head(3)
```

```python
# Since farm name can be extremely specific, probably best to remove this one for now
merged_cleaned = merged_cleaned.drop('Farm.Name', axis=1)
merged_cleaned.head(3)
```

```python
merged_cleaned.columns
```

```python
# We'll also get rid of Lot number, ICO.Number, Company, Altitude, etc
merged_cleaned = merged_cleaned.drop(['Lot.Number','ICO.Number','Company','Altitude','Expiration','Grading.Date'
                                      'Certification.Body','Certification.Address','Certification.Contact'], axis=1)
```

```python
# most of the data seems to be measured in meters, but some are in feet. Looks like altitude low, high, and mean all get converted to meters though
merged_cleaned['unit_of_measurement'].value_counts()
```

```python
# just keeping high and low altitude in meters for now
merged_cleaned = merged_cleaned.drop(['unit_of_measurement','altitude_mean_meters'],axis=1)
```
